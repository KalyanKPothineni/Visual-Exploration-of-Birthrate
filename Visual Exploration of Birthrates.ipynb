```python
import pandas as pd

# Load the birth-rate.csv file
birth_rate_data = pd.read_csv('birth-rate.csv')
birth_rate_data.head()
```

```python
# Load the crimeratesbystate-formatted.csv file
crime_rate_data = pd.read_csv('crimeratesbystate-formatted.csv')
crime_rate_data.head()
```

```python
# Load the education.csv file
education_data = pd.read_csv('education.csv')
education_data.head()
```

```python
import matplotlib.pyplot as plt

# Histogram for birth rates in 2008
plt.figure(figsize=(10, 6))
plt.hist(birth_rate_data['2008'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Birth Rates in 2008')
plt.xlabel('Birth Rate per 1000 People')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.show()
```

```python
# Box Plot for robbery rates across different states
plt.figure(figsize=(10, 6))
plt.boxplot(crime_rate_data['robbery'], vert=False, patch_artist=True, flierprops=dict(marker='o', color='red', markersize=5))
plt.title('Box Plot of Robbery Rates Across States')
plt.xlabel('Robbery Rate per 100,000 People')
plt.grid(True)
plt.show()
```

```python
# Simulate a bullet chart using bar charts to compare actual dropout rates to a target
# Setting a target dropout rate of 6%
target_dropout_rate = 6

plt.figure(figsize=(14, 8))
bars = plt.bar(education_data['state'], education_data['dropout_rate'], color='gray')
target = plt.axhline(y=target_dropout_rate, color='r', linestyle='--', linewidth=2)

# Highlight bars that exceed the target
for bar in bars:
    if bar.get_height() > target_dropout_rate:
        bar.set_color('tomato')

plt.title('Dropout Rates by State with Target Line (6%)')
plt.xlabel('State')
plt.xticks(rotation=90)
plt.ylabel('Dropout Rate (%)')
plt.legend([target], ['Target Dropout Rate'])
plt.tight_layout()
plt.show()
```

```python
# Creating a bubble chart to visualize SAT scores and dropout rates
plt.figure(figsize=(12, 8))
bubble_sizes = education_data['dropout_rate'] * 10  # Scale factor for better visualization

scatter = plt.scatter(education_data['reading'], education_data['math'],
                      s=bubble_sizes, color='cyan', alpha=0.6, edgecolors='black', linewidth=1)

plt.title('Bubble Chart: SAT Reading vs. Math Scores with Dropout Rates')
plt.xlabel('SAT Reading Scores')
plt.ylabel('SAT Math Scores')
plt.grid(True)

# Adding a colorbar to represent dropout rates
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = plt.legend(handles, labels, loc="upper right", title="Dropout Rate (%)")

plt.show()
```

```python
# Assigning regions to states for color coding in the bubble chart
regions = {
    'Northeast': ['Connecticut', 'Massachusetts', 'Maine', 'New Hampshire', 'Rhode Island', 'Vermont', 'New York', 'Pennsylvania', 'New Jersey'],
    'Midwest': ['North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri', 'Wisconsin', 'Illinois', 'Michigan', 'Indiana', 'Ohio'],
    'South': ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana'],
    'West': ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']
}

# Mapping each state to its region
education_data['region'] = education_data['state'].apply(lambda x: next((key for key, value in regions.items() if x in value), 'Other'))

# Mapping colors
region_colors = {'Northeast': 'blue', 'Midwest': 'green', 'South': 'red', 'West': 'purple', 'Other': 'gray'}

# Creating a bubble chart with regions color coded
plt.figure(figsize=(12, 8))
for region, group in education_data.groupby('region'):
    plt.scatter(group['reading'], group['math'], s=group['dropout_rate'] * 10,
                color=region_colors[region], alpha=0.6, edgecolors='black', label=region, linewidth=1)

plt.title('Bubble Chart: SAT Reading vs. Math Scores with Dropout Rates by Region')
plt.xlabel('SAT Reading Scores')
plt.ylabel('SAT Math Scores')
plt.grid(True)
plt.legend(title='Region')
plt.show()
```
